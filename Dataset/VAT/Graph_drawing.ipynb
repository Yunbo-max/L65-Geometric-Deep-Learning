```python
import pandas as pd

# Load the DataFrame from the CSV file
df = pd.read_csv('./Data/pp_dep_2017_2019.csv', delimiter=';')

# Load the DataFrame containing the harmonized system information
df2 = pd.read_csv('./Data/harmonized-system.csv', delimiter=',')

# Extract the first two digits from the "HS6_i" and "HS6_j" columns
df['label_i'] = df['HS6_i'].astype(str).str[:2]
df['label_j'] = df['HS6_j'].astype(str).str[:2]

# Filter df2 to only include rows with two-digit HS codes
df2_two_digits = df2[df2['hscode'].astype(str).str.len() == 2]

# Merge df with df2_two_digits to add labels based on matching first two digits
df = pd.merge(df, df2_two_digits[['hscode', 'description']], how='left', left_on='label_i', right_on='hscode')
df.rename(columns={'description': 'label_i_desc'}, inplace=True)
df = pd.merge(df, df2_two_digits[['hscode', 'description']], how='left', left_on='label_j', right_on='hscode')
df.rename(columns={'description': 'label_j_desc'}, inplace=True)

# Drop unnecessary columns
df.drop(['hscode_x', 'hscode_y'], axis=1, inplace=True)

# Display the updated DataFrame
df
```

```python
# Save the DataFrame to a new CSV file
df.to_csv('./Data/label_pp_dep_2017_2019.csv', index=False)
```
